```python
import proj_helpers.FinObj as fin
import numpy as np
import proj_helpers.riskEngine as risk
import matplotlib.pyplot as pplot
import time
```

```python
#################################################################
#################################################################
#########     Demo to Plot Various Payoff Grpahs      ###########
#################################################################
#################################################################
```

```python
#Stock Parameters
spot = 100
vol = 0.2 #vol per year
rate = 0.1 #Rate in years
div = 0.0
strike=100
payoff=fin.amPutVec #Define option payoff (toggles the first graph)
time_s = 10 #Total number of time steps in t_units
t_unit = 10

#Share object
share1 = fin.Stock(spot,vol,t_unit,rate)

#Option Parameters
myOption = fin.Option(share1,payoff,time_s,strike,t_unit)
```

```python
spots = np.linspace(80,120,100)
payoffs = fin.amPutVec(spots,myOption)
```

```python
################################################
######## Plot Vanilla Put Payoff ###############
################################################
pplot.plot(spots,payoffs)
pplot.xlabel('Spot Price')
pplot.ylabel('Payoff Value')
pplot.show()
```

```python
################################################
######## Plot Vanilla Digital Payoff ###########
################################################
payoffs = fin.amPutDigVec(spots,myOption)
pplot.plot(spots,payoffs)
pplot.xlabel('Spot Price')
pplot.ylabel('Payoff Value')
pplot.show()
```

```python
################################################
######## Plot Wedding Cake Payoff ##############
################################################
payoffs = fin.birthdayVec(spots,myOption)
pplot.plot(spots,payoffs)
pplot.xlabel('Spot Price')
pplot.ylabel('Payoff Value')
pplot.show()
```

```python
################################################
######## Plot Double Digital Payoff ############
################################################

payoffs = fin.wedgeVec(spots,myOption)
pplot.plot(spots,payoffs)
pplot.xlabel('Spot Price')
pplot.ylabel('Payoff Value')
pplot.show()
```
